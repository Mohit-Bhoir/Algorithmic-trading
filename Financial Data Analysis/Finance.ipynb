**SIMPLE RETURNS VS LOG RETURNS (PART 1)**

```python
df = pd.DataFrame(data = [100,50,90], columns=['Price'])
df 
```

```python
periods = df.SR.count()
periods
```

**Arithmetic mean of simple returns can be misleading**

```python
mean_sr = df.SR.mean()
mean_sr
```

```python
100 * ( 1 + mean_sr ) ** periods
```

**We should use compound simple returns / geometric mean or even better**

```python
geo_mean = ( 1 + df.SR ).prod() ** (1/periods) - 1
geo_mean
```

```python
100 * (1 + geo_mean) ** periods
```

**Log returns are additive over time!**

```python
sum_lr = df.LR.sum()
sum_lr
```

```python
100 * np.exp(sum_lr)
```

```python
mean_lr = df.LR.mean()
mean_lr
```

```python
100 * np.exp(mean_lr * periods)
```

*Log returns are additive over time. Simple returns are not additive over time (but they can be multiplied/compounded)*

*Testing the normality of MSFT Returns based on the sample (Oct 2014 to May 2021)-> Hypothesis test with H0 Hypothesis: MSFT Returns (full population) follow a normal distribution*

```python
z_stat, p_value = stats.normaltest(msft.Log_Returns.dropna())
```

```python
z_stat #High value -> reject H0
```

```python
p_value #Low value (Closest to zero) -> reject H0
```

```python
round(p_value, 10)
```

*Assuming that MSFT Returns (generally) follow a normal distribution, there is 0% probability that we get that extreme outcomes in a sample*

*MSFT Returns doesn't follow a normal distribution as they exhibit fat tails. Extreme events are not reflected in mean-variance analysis. The standard deviation of returns underestimates true risk*

*How will mean-variance analysis change with smoothed data?*

```python
freqs = ['YE','QE','ME','W-Fri','D']
periods = [1,4,12,52,252]
ann_mean = []
ann_std = []
```

```python
for i in range(5):
    resamp = msft.Price.resample(freqs[i]).last() #RESMAPLING
    ann_mean.append(np.log(resamp / resamp.shift(1)).mean() * periods[i]) #Annualized Mean Log Return
    ann_std.append(np.log(resamp / resamp.shift(1)).std() * np.sqrt(periods[i])) #Annualized Standard Deviation of Log Returns
```

```python
ann_mean
```

```python
ann_std
```

```python
summary = pd.DataFrame(data = {'Mean': ann_mean, 'Std': ann_std}, index=freqs)
summary
```

```python
summary.plot(kind='scatter', x='Std', y='Mean', figsize=(10,6), fontsize=13, title='Risk vs Reward with Resampled Data')
for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i,'Std']+0.00005, summary.loc[i,'Mean']+0.00005), fontsize=12)
plt.ylim(0,0.3)
plt.xlabel('Risk (Standard Deviation)', fontsize=13)
plt.ylabel('Reward (Mean Return)', fontsize=13)
plt.show()
```

**MARGIN TRADING & LEVERED RETURNS**

*A trader buys a stock (stock price:100) on margin 50%. After one day the price increases to 110. Calculate unlevered return and levered return*

```python
P0 = 100   
P1 = 110
leverage = 2 #2:1 leverage
margin = P0 / leverage
```

```python
margin
```

```python
unlev_return = (P1 - P0) / P0 #Simple Unlevered Return
unlev_return
```

```python
lev_return = (P1 - P0) / margin #Simple Levered Return
lev_return
```

```python
lev_return == unlev_return * leverage
```

```python
unlev_return = np.log((P1-P0)/P0 + 1) #Log Unlevered Return
unlev_return
```

```python
lev_return = np.log((P1 - P0) / margin + 1) #Log Levered Return
lev_return
```

```python
lev_return == unlev_return * leverage #This relationship does not holds true for log returns as well
```

*To calculate levered returns, dont multiply leverage with log returns*
