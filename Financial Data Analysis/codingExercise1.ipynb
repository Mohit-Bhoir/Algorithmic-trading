```python
import pandas as pd 
import yfinance as yf
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
plt.style.use('seaborn-v0_8')
```

```python
start = "2015-01-02"
end = "2020-12-31"
symbol = ['GE','RELIANCE.NS']
```

```python
df = yf.download(symbol, start=start, end=end)
df.head()
```

```python
close = df.Close.copy()
close
```

```python
close.dropna().plot(figsize=(12,6), fontsize = 13, title='Closing Prices')
plt.legend(fontsize=13)
plt.show()
```

```python
norm = close.div(close.iloc[0]).mul(1)
norm
```

```python
norm.dropna().plot(figsize=(12,6), fontsize = 13, title='Normalized Closing Prices')
plt.legend(fontsize=13)
plt.show()
```
